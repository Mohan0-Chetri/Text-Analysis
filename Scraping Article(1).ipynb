```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
```

```python
browser = webdriver.Chrome(r"C:\Users\mchet\Downloads\chromedriver_win32\chromedriver.exe")
```

```python
from selenium.webdriver.common.keys import Keys
import time
```

```python
df=pd.read_excel(r"C:\Users\mchet\Downloads\Input (1).xlsx")
```

```python
urls = df['URL'].tolist() #Converting the columns into list
len(urls)
```

```python
#scrape elements
cp=[]
for url in urls:
    browser.get(url)
    content = browser.page_source
    soup = BeautifulSoup(content, "lxml")
    cus_list=[]
    for item in soup.find_all("div", class_="td-post-content"):
        data_str = item.get_text()
        cus_list.append(data_str)
        d=cus_list
    cp.append(d)
        
```

```python
#Cleaning the newline tags '/n'
```

```python
cp[0]
```

```python
def cl(text):
    return text.replace('\n',"")
```

```python
clean=[]
for i in cp:
    for j in i:
        z=j.replace('\n',"")
        clean.append(z.replace('\n',""))
        
        
```

```python
clean[0]
```

```python
data=pd.DataFrame()
data['URL_ID']=(df['URL_ID']).astype(int)
data['Article']=clean
```
